```python
testing = combine[['biccontenttypename','addedtocart','trafficcount','videocount',
                   'grs1month', 'grs2month', 'grs3month', 'grs12month', 'expectedgrs','modulecount', 'clname', 'mkcname']]
```

```python
testing.head()
```

```python
testing['CustomerExp'] = ''
testing['CustomerExp'] = testing['addedtocart'] / testing['trafficcount']
```

```python
testing['Revenue'] = ''
testing['Revenue'] = testing['grs12month'] + testing['expectedgrs']
```

```python
testing
```

```python
anova_data = testing[['biccontenttypename', 'clname', 'CustomerExp', 'Revenue']]
```

```python
anova_data
```

```python
anova_data.isnull().sum()
```

```python
anova_nodup = anova_data[anova_data['CustomerExp'].notnull()]
anova_nodup = anova_nodup[anova_nodup['clname'].notnull()]
```

```python
anova_nodup.to_csv('anova_nodup.csv', index=False)
anova_data.to_csv('anova_full.csv', index=False)
```
